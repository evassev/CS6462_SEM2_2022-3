### Supervised Learning

<b><i>Learn by identifying patterns in data that is already labeled.</i></b>

<ul>
  <li><b>Classification</b></li>
    <ul>
      <li>Target: predict (or classify) discrete values (male or female, heads or tails); map input to predefined categories</li> 
      <li><b>Binary Classification</b></li>
        <ul>  
          <li>Target: Binary values [0,1]</li> 
          <li>Applications: fraud detection, medical diagnostics</li>
        </ul>    
      <li><b>Milti-class Classification</b></li>
        <ul>  
          <li>Target: Multiple values [0,1,2,...,n] - 3 or more classes of data </li> 
          <li>Applications: image recognition, medical diagnostics, staff classification, personalized advertising</li>
        </ul>
    </ul>
  <li><b>Regression</b></li>  
    <ul>  
      <li>Target: predict continuous values (price, salary, age); map input to a contineous value (e.g., integer)</li> 
      <li>Applications: product sales prediction, weather forecasting, market forecasting</li>
    </ul>    
</ul>    

### Unsupervised Learning

<b><i>Automatically uncover and create data labels by 1)detecting emerging properties on the entire dataset; and 2) constructing patterns out of these properties.</i></b>

<ul>
  <li><b>Clustering</b></li>
    <ul>
      <li>Target: group data into different clusters based on similar features</li> 
      <li>Applications: product recommendations, customer segmentation, targeted marketing, medical diagnostics</li>
    </ul>
  <li><b>Dimensionality Reduction</b></li>  
    <ul>  
      <li>Target: reduce the number of input variables in a dataset</li> 
      <li>Applications: visualization, natural language processing, data structure discovery, gene sequencing</li>
    </ul>    
</ul>    

### Reinforcement Learning

<b><i>Learning through trial and error. Best when the desired outcome is known but the exact path to achieving it is not known.</i></b>

The computer employs trial and error to come up with a solution to the problem. 
To get the machine to do what the programmer wants, the artificial 
intelligence 

<ul>
  <li>Target: Get either rewards or penalties for the perfomed actions. The goal is to maximize the total reward.</li> 
  <li>Applications: game AI, self-driving cars, robotics, customer service routing</li>
</ul>

### 3) Analyze the Target Variable

<i>The **target variable** of a dataset is the feature of a dataset about which we want to gain a deeper understanding. This is the dependable variable.</i><br>


```python
df.dtypes
```

</i>Our target variable is the diamond cut or **cut**.</i>

<i>The **cut** values are not numbers, and therefore, they are not appropriate for our ML model. Thus, we convert this column to a numeric value.</i>

```python
class_mapper = {'Ideal':1,'Premium':2,'Very Good':3,'Good':4,'Fair':5}
df['cut']=df['cut'].replace(class_mapper)
```

<i>We plot the 'carat' against the target.</i>

<i>There appears to be only a slight difference comparing the **carat** values and the different **cut** values. Probably, there is no correlation.</i>
<br><br>
<i>We plot the 'depth' against the target.</i>

```python
df.plot.scatter(y='depth',x='cut')
```

<i>There appears to be some difference comparing the **depth** values and the different **cut** values. Probably, there is a correlation.</i><br><br>
<i>We use the **corr** function to create a correlation matrix for the entire dataset.</i>

```python
corr_matrix = df.corr()
corr_matrix["cut"].sort_values(ascending=False)
```

</i>We can conclude that the 'table' and 'cut' features are highly correlated. Note that the 'table' feature is the **diamond table facet** - usually the largest facet on the diamond.</i><br><br>

## Regression ML

<i>Predict *continuous values* (carat, depth, table, price, etc.)</i>  

```python
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import train_test_split
```

<i>Both **color** and **clarity** are not numerical, and therefore, they are not appropriate for our ML model. Thus, we convert these two columns to numerical.</i>

```python
df['color'].value_counts()
#df.groupby(['color']).count()
```

```python
class_mapper_color = {'G':1,'E':2,'F':3,'H':4,'D':5,'I':6, 'J':7}
df['color']=df['color'].replace(class_mapper_color)
```

```python
df['clarity'].value_counts()
```

```python
class_mapper_clarity = {'SI1':1,'VS2':2,'SI2':3,'VS1':4,'VVS2':5,'VVS1':6, 'IF':7, 'I1':8}
df['clarity']=df['clarity'].replace(class_mapper_clarity)
```

We are going to train our ML system to predict values of $z$ (one of the diamond dimensions).
Here, $z$ is our observation.

We extract:
<li>$X$ - a dataset of all the variables but $z$</li>
<li>$Y$ - vector that contains all the values of $z$</li> 

```python
Y = df.iloc[:,9]
X = df.iloc[:,:9]
```

We slice the dataset: 80% for training and 20% for testing.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

We create an instance of the RandomForestRegressor ML model and train (we use the $fit()$ method).

```python
rf_regressor = RandomForestRegressor (n_estimators=100)
rf_regressor.fit(X_train, Y_train)
```

We test by using the $predict()$ method.

```python
Y_pred = rf_regressor.predict(X_test)
```

<i>Compute the linear regression score. 1 is maximum.</i>

```python
from sklearn import metrics
print(metrics.r2_score(Y_test,Y_pred))
```
